def initialize_inventory(products):

    return inventory

````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_qty = False
        while not valid_qty:
            try:
                qty = int(input(f"Enter the quantity of {product}s available: "))
                if qty < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_qty = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = qty
    return inventory

def get_cust_ord(inventory):
    try:
        num_orders = int(input("Enter the number of customer orders: "))
        if num_orders < 0:
            raise ValueError("Invalid number of orders! Please enter a non-negative value.")
    except ValueError as error:
        print(f"Error: {error}")
        return []

    cust_ord = [input("Enter the name of a product that a customer wants to order: ") for _ in range(num_orders) if input("Enter the name of a product that a customer wants to order: ") in inventory and inventory[input("Enter the name of a product that a customer wants to order: ")] > 0]
    return cust_ord

def calculate_total_price(cust_ord):
    try:
        total_price = sum([float(input(f"Enter the price of {product}: ")) for product in cust_ord if float(input(f"Enter the price of {product}: ")) >= 0])
        return total_price
    except ValueError as error:
        print(f"Error: {error}")
        return 0

def update_inventory(inventory, cust_ord):
    for product in cust_ord:
        inventory[product] -= 1
    inventory = {product: qty for product, qty in inventory.items() if qty > 0}
    return inventory

````

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]  # Sample list of products
inventory = initialize_inventory(products)  # Initialize inventory

cust_ord = get_cust_ord(inventory)  # Get customer orders
total_price = calculate_total_price(cust_ord)  # Calculate total price
inventory = update_inventory(inventory, cust_ord)  # Update inventory

print("Order Statistics:")
print(f"Total Products Ordered: {len(cust_ord)}")
print(f"Percentage of Unique Products Ordered: {len(set(cust_ord))/len(products)*100}%\n")
print("Updated Inventory:")
for product, qty in inventory.items():
    print(f"{product}: {qty}")
print(f"Total Price: {total_price}")
````
